```python
#step-2 Normalization
def normalization(X):
    u = X.mean(axis = 0)
    std = X.std(axis=0)

    return (X-u)/std
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Hypothesis function to predict values using X (features) and theta (parameters)
def hypothesis(X, theta):
    return np.dot(X, theta)

# Loss function to calculate mean squared error (MSE)
def loss(X, y, theta):
    yp = hypothesis(X, theta)  # predicted values
    error = np.mean((y - yp) ** 2)  # MSE calculation
    return error

# Gradient function to compute the derivative of the loss function w.r.t. theta
def gradient(X, y, theta):
    yp = hypothesis(X, theta)  # predicted values
    grad = np.dot(X.T, (yp - y))  # gradient computation
    m = X.shape[0]  # number of training examples
    return grad / m  # average gradient

# Train function to perform gradient descent optimization
def train(X, y, learning_rate=0.1, max_iters=100):
    n = X.shape[1]  # number of features
    theta = np.random.randn(n)  # initialize theta randomly
    error_list = []  # to keep track of error at each iteration

    for i in range(max_iters):
        e = loss(X, y, theta)  # compute current loss
        error_list.append(e)  # store the loss
        grad = gradient(X, y, theta)  # compute the gradient
        theta = theta - learning_rate * grad  # update theta
        
    plt.plot(error_list)  # plot the loss over iterations
    plt.xlabel('Iterations')
    plt.ylabel('Loss')
    plt.title('Training Error over Time')
    plt.show()  # Ensure the plot is displayed
    
    return theta  # return the optimized parameters


```
